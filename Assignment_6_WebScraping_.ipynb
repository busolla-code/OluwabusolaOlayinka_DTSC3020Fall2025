### 2) Common Imports & Polite Headers

```python
# Common Imports & Polite Headers
import re, sys, pandas as pd, requests
from bs4 import BeautifulSoup
HEADERS = {"User-Agent": (
    "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
    "(KHTML, like Gecko) Chrome/122.0 Safari/537.36")}
def fetch_html(url: str, timeout: int = 20) -> str:
    r = requests.get(url, headers=HEADERS, timeout=timeout)
    r.raise_for_status()
    return r.text
def flatten_headers(df: pd.DataFrame) -> pd.DataFrame:
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = [" ".join([str(x) for x in tup if str(x)!="nan"]).strip()
                      for tup in df.columns.values]
    else:
        df.columns = [str(c).strip() for c in df.columns]
    return df
print("Common helpers loaded.")
```

```python
# Q2 — Write your answer here

def q2_parse_items(html: str) -> pd.DataFrame:
    soup = BeautifulSoup(html, "lxml")
    items = []
    rows = soup.select(".athing")
    for row in rows:
        rank = row.select_one(".rank")
        title = row.select_one(".titleline a")
        subtext = row.find_next_sibling("tr").select_one(".subtext")
        points = subtext.select_one(".score") if subtext else None
        comments = None
        if subtext:
            links = subtext.select("a")
            for link in links:
                if "comment" in link.text:
                    comments = link
                    break
        items.append({
            "rank": rank.get_text(strip=True) if rank else "",
            "title": title.get_text(strip=True) if title else "",
            "link": title["href"] if title else "",
            "points": points.get_text(strip=True) if points else "",
            "comments": comments.get_text(strip=True) if comments else ""
        })
    return pd.DataFrame(items)

def q2_clean(df: pd.DataFrame) -> pd.DataFrame:
    df = df.fillna("")
    df["rank"] = df["rank"].str.extract(r"(\d+)").fillna("0").astype(int)
    df["points"] = df["points"].str.extract(r"(\d+)").fillna("0").astype(int)
    df["comments"] = df["comments"].str.extract(r"(\d+)").fillna("0").astype(int)
    return df

def q2_sort_top(df: pd.DataFrame, top: int = 15) -> pd.DataFrame:
    df = df.sort_values(by="points", ascending=False)
    return df.head(top)

html = fetch_html("https://news.ycombinator.com/")
df = q2_parse_items(html)
df = q2_clean(df)
top15 = q2_sort_top(df)
top15.to_csv("data_q2.csv", index=False)
top15
```
