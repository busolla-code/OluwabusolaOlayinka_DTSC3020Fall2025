## 0) Setup

```python
# Run it first.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style='whitegrid')
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()  # preview
```

## 1–3) Scatter plot with color by species
Create a scatter plot with:
- **x =** `bill_length_mm`
- **y =** `bill_depth_mm`
- Color points by **species**

Then add a short comment (2–3 sentences) describing the pattern(s) you see.

```python
# TODO: Create the scatter plot here

import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

plt.figure(figsize=(8,6))
sns.scatterplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", palette="deep")
plt.title("Penguin Bill Dimensions by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.legend(title="Species")
plt.show()
```

**Your interpretation (2–3 sentences):**

Adelie's have shorter bill lengths and moderately greater bill depths. Chinstrap's have bill lengths that are longer than Adelie's yet shorter than the Gentoo. The Chinstrap bill depth is about the same in comparison to the Adelie. The Gentoo's have about the same bill depths as the Adelie's and generaly similar bill lengths as the Chinstrap.

## 4) Histogram with two different bin sizes
Plot a histogram of `bill_length_mm` **twice**:
- Once with **10 bins**
- Once with **30 bins**

Then write 1–2 sentences comparing what changes when you change the number of bins.

```python
# TODO: Histogram with 10 bins

plt.figure(figsize=(8,6))
sns.histplot(data=penguins, x='bill_length_mm', bins=10, color='blue', kde=True)
plt.title('Histogram of Bill Length (10 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')
plt.show()
```

```python
# TODO: Histogram with 30 bins
plt.figure(figsize=(8,6))
sns.histplot(data=penguins, x='bill_length_mm', bins=30, color='pink', kde=True)
plt.title('Histogram of Bill Length (30 bins)')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Count')
plt.show()
```

**Your comparison (1–2 sentences):**

With less bins, the Histogram looks easier to read and less complicated, while the Histogram with 30 bins is more intricate and detailed.

## 5) Boxplot of body mass by species
Create a **boxplot** showing `body_mass_g` grouped by `species`.
Then write 1–2 sentences summarizing which species are heavier/lighter on average.

```python
# TODO: Boxplot of body mass by species

plt.figure(figsize=(5,6))
sns.boxplot(data=penguins, x='species', y='body_mass_g', palette='pastel')
plt.title('Body Mass by Penguin Species')
plt.xlabel('Species')
plt.ylabel('Body Mass (g)')
plt.show()
```

**Your summary (1–2 sentences):**

The Adelie and Chinstrap are very similar in body mass. They are much lighter in comparison to the Gentoo. The Gentoo penguin is heaviest on average.
